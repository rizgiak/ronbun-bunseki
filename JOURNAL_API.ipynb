```python
from journal_api import crossref_api
from journal_api import semanticscholar_api
from journal_api import arxiv_api
from journal_api import googlebook_api
from IPython.display import JSON
```

```python
max_pos = 3

def dic_library(title, pos):
    res = None
    if pos == 1:
        res = crossref_api.search(title)
    elif pos == 2:
        res = arxiv_api.search(title)
    elif pos == 3:
        res = googlebook_api.search(title)
    else:
        res = semanticscholar_api.search(title) 
    
    return res

def _remake_result(result, pos):
    if (result == None):
        return None
    dic = ["abstract", "year", "fieldsOfStudy", "authors", "references"]
    l = [result[dic[0]] == "", result[dic[1]] == "", result[dic[2]] == "", len(result[dic[3]]) == 0, len(result[dic[4]]) == 0]

    if any(l): # data not completed
        for i in range(pos + 1, max_pos + 1):
            rc = dic_library(result["title"], i)
            if rc != None:
                r = [rc["abstract"] == "", rc["year"] == "", rc["fieldsOfStudy"] == "", len(rc["authors"]) == 0, len(rc["references"]) == 0]
                for index, value in enumerate(l):
                    if(value == True and r[index] == False):
                        l[index] = False
                        result[dic[index]] = rc[dic[index]]
                if not any(l):
                    break

    return result

def _fix_references(result, year):
    if (result == None):
        return None
    
    if len(result["references"]) > 0:
        refs = []
        for ti in (result["references"]):
            for i in range(max_pos + 1):
                res = dic_library(ti, i)
                if(res != None):
                    if res["year"] >= year:
                        refs.append(ti)
                    break;
        result["references"] = refs
    return result


def search(title, start_year = 2010):
    res = None
    for i in range(max_pos + 1):
        res = dic_library(title, i)
        if (res == None or res["year"] < start_year): # year limitation
            return None
        res = _remake_result(res, i)
        res = _fix_references(res, start_year)
        if(res != None):
            break;
    return res
```

```python
title = "Quantum Dynamical Approach to Predicting the Optical Pumping Threshold for Lasing in Organic Materials!!!??"
search(title, 2020)
```
